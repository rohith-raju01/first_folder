```python
def checkingwhile(m):  
    x=m

    while x>=100:
        print(x)
        x=x**2
        
checkingwhile(2)
```

```python
a=1
sum=0
while(a<=100):
    sum=sum+a
    a=a+1
print(sum)
```

```python
def sum_natural(m):
    s=0
    for i in range(1,m+1):
        s=s+i
        print(s)
    
sum_natural(300)
```

```python
def oddnumbers(m):
    for i in range(3,m+2,2):
        print(i)
oddnumbers(99)
```
